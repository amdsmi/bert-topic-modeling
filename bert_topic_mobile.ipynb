```python
import pandas as pd
data=pd.read_csv('/content/drive/MyDrive/mobile_re.csv')
```

```python
data['date'] = pd.to_datetime(data['date'])
```

```python
data=data[['body','rating','date']]
```

```python
data=data.dropna()
```

```python
data_pos=data[data.rating>3]
data_neg=data[data.rating<3]
data_ntr=data[data.rating==3]
data_pos['sent']='pos'
data_neg['sent']='neg'
data_ntr['sent']='ntr'
```

```python
data=pd.concat([data_pos,data_neg,data_ntr])
```

```python
data
```

```python
docs=data.body.values
classes=data.sent.values
timestamps=data.date.values
```

```python
from bertopic import BERTopic
```

```python
topic_model = BERTopic(verbose=True)
topics, probs = topic_model.fit_transform(docs)
```

```python
topics_over_time = topic_model.topics_over_time(docs, topics, timestamps,
                                                global_tuning=True, evolution_tuning=True)
```

```python
topic_model.visualize_topics_over_time(topics_over_time, top_n_topics=50)
```

```python
topics_per_class = topic_model.topics_per_class(docs, topics, classes=classes)
```

```python
topic_model.visualize_topics_per_class(topics_per_class,top_n_topics=50)
```

```python
topic_model.visualize_hierarchy(top_n_topics=50)
```

```python
topic_model.visualize_heatmap(top_n_topics=50)
```

```python
topic_model.visualize_barchart(top_n_topics=16)
```

```python
topic_model.visualize_topics(top_n_topics=50)
```
